CNN. DATASET FASHION MINIST

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
```

```python
(train_imagenes, train_labels),(test_imagenes, test_labels)= fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(40):
    image=train_imagenes[i].reshape(28,28)
    plt.imshow(image,cmap='gray')
    plt.show()
```

```python
train_imagenes=train_imagenes/255.0
test_imagenes=test_imagenes/255.0
```

```python
train_imagenes=train_imagenes.reshape((-1,28,28,1))
test_imagenes=test_imagenes.reshape((-1,28,28,1))
```

```python
model= Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(2,2),   #matriz de 2x2 para reducir
    Dropout(0.25),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Dropout(0.25),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics='accuracy')
```

```python
history=model.fit(train_imagenes, train_labels, epochs=3, validation_split=0.2)
```

```python
test_loss, test_acc=model.evaluate(test_imagenes, test_labels)
print('\nTest accuracy:', test_acc)
```

```python
import matplotlib.pyplot as plt
history=model.fit(train_imagenes, train_labels, epochs=3, validation_split=0.2)
```

```python
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```
